# Open Excel data

<ul>
    <li> Data in Excel can be opened to a dataframe using command <strong>read_excel</strong> of the Pandas library </li>
    <li>If your file locates in the same directory as your code and data starts from the first cell of your Excel file, it is enough to add the file name as argument. </li>
    <li>If your file locates in another directory or is an Internet resource, you should add the appropriate path; e.g. 'C:/Users/Andrea/Documents/mydata.xlsx'. </li> 
    <li>After your is opened in the dataframe, the first five rows can be viewed with the command <strong>df.head()</strong> and, similarly, the last five rows using the command <strong>df.tail()</strong>. </li>
    <li>If your data does not start from the first cell of your workbook, further arguments <strong>sheet_name</strong>, <strong>usecols</strong> and <strong>skiprows</strong> may be needed.
    <ul>
        <li>For instance, in order to open data from your workbook sheet called <em>Sheet2</em> and its columns B and C but ignoring the first row, you should type the following:<br /> <em>df = pd.read_excel('mydata.xlsx', sheet_name = 'Sheet2', usecols = 'B, C', skiprows = 1)</em> </li>
    </ul>
    </li>
    <li>In case it is more unwieldy situation, you may consider cleaning your data in Excel <em>before</em> opening it with the command <strong>read_excel</strong>. </li>
<li> For further information about the command <strong>read_excel</strong>, see 
    <a href="https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.read_excel.html">https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.read_excel.html</a>
</ul>

```python
import pandas as pd
```
