Variables that contain only numerical values are either integers (int) or floating-point numbers (float). Floating-point numbers can be practically be considered as decimal numbers.

```python
### Numbers of missing values

df.isnull().sum()

### Compare the command above and the one without sum()
```

Using the command <strong>unique()</strong> it is easy to find false or suspicious values.

The command <strong>unique()</strong> gives a list of unique values of variable. 

For instance, the following example shows us that the variable colleagues (satisfaction with colleagues) has values 2, 3, ,4 and 5 and at least one missing value (nan).

```python
print(df['colleagues'].unique())

### Again, compare this without unique()
```

For loops are traditional in programming. See more about for loops in Python at https://wiki.python.org/moin/ForLoop

```python
### Using for loops in order to go through all the variables of the data

for var in df:
    print(var, df[var].unique())
```

```python
### List n biggest values of a column

df.nlargest(n = 3, columns = 'salary')
```

```python
### Correspondingly, n smallest values

df.nsmallest(n = 3, columns = 'salary')
```

The command <strong>value_counts()</strong> counts frequencies (i.e., the number of occurrencies) of unique values. 

In the following example frequencies are counted to all variables except the running number (number).

The command <strong>drop()</strong> omits rows as default (axis = 0). However, in the following example columns are omitted in the calculation (axis = 1).

```python
for var in df.drop('number', axis = 1):
    print(df[var].value_counts())
```
